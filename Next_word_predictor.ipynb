```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import LSTM,Embedding,Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
import numpy as np
import re
```

# Text cleaning

```python

def clean_text(text):

    # convert to lowercase
    text = text.lower()

    # remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
cleaned_faqs=clean_text(faqs)
```

```python
cleaned_faqs
```

## converting into list for the word_index to word level. without it [] happens character level

```python
in_list=[cleaned_faqs]
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(in_list)
```

# assign unique interger values to each unique word for the model.

```python
tokenizer.word_index
```

# each word appearacne count

```python
tokenizer.word_counts
```

# assigning the `integer numbers(word_index)` to each senetnece and also organizing into the increasing order to each sentence for the making input-label structure'

```python
input_sequences=[]
for sentence in faqs.split('\n'):
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])

```

```python
len(input_sequences)
```

```python
max_lengh=max([len(x) for x in input_sequences])
```

```python
max_lengh
```

`pre->' so that we can extract the last word_idex of each list for label.

'truncating->` for the truncate the word which is greater than the max_lengh for gpu compitability.

`max_len->` max word allowed in an sentence.

```python
paded_sequences=pad_sequences(input_sequences,maxlen=max_lengh,padding='pre',truncating='pre')
```

```python
paded_sequences
```

# training data is ready in the from of input->label.

```python
x=paded_sequences[:,:-1]
y=paded_sequences[:,-1]
```

```python
y
```

```python
x.shape
```

`to_categorical` convert the label into ohe encoded

```python
# +1 cuz word index start with 1 and 0 is preserved for the padding.
vocab_size=len(tokenizer.word_index)+1
```

```python

y=to_categorical(y,num_classes=vocab_size)
```

```python
y.shape
```

```python

model = Sequential()

model.add(Embedding(vocab_size, 100, input_length=24))

model.add(LSTM(150, return_sequences=True))

model.add(LSTM(140))

model.add(Dense(vocab_size, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=100)
```

```python
import time
text = "What is training data"

for i in range(15):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
